### Example



```python
# Setting
from collections import deque
N = 6
graph = [[], [2, 3], [4, 5], [4, 5], [], [6], []]
```

```python
# In degree
in_deg = [0]*(N+1)
for g in graph:
    for end in g:
        in_deg[end] += 1
```

```python
# Algorithm
result = []
que = deque()
for node in range(1, N+1):
    if in_deg[node] ==0:
        que.append(node)

while que:
    now = que.popleft()
    result.append(now)
    for adj in graph[now]:
        in_deg[adj] -= 1
        if in_deg[adj]==0:
            que.append(adj)
```

```python
# Print result
for i, r in enumerate(result):
    if i == N-1:
          print(f'{r}')
    else:
        print(f'{r}', end = '--->')
```
