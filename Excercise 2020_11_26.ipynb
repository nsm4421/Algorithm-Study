### 볼링공 고르기

> 문제

1번 부터 N번까지 볼링공이 있는데 각 공의 무게는 1에서 M까지의 자연수로 주어진다.

2개의 공을 고르는데 서로의 공의 무게는 달라야 한다.

가능한 조합의 수는?


> 예시

예를 들어 N =3, M = 2이고, 공의 무게가 1,2,2 로 주어진다고 하자.

그럼 (1번, 2번), (1번, 3번)의 조합이 가능하므로 2가지이다.

> 입력

첫번째 줄에는 공의 개수 N과 공의 최대 무게 M이 주어진다.

두번째 줄에는 N개의 공들의 무게가 차례로 주어진다.

```python
N, M = map(int, input().split())
weights = list(map(int, input().split()))
```

```python
cnt = 0
for i in weights:
    for j in weights:
        if i!=j:
            cnt += 1
print(f'{cnt//2}')
```

### 럭키 스트레이트

> 조건

현재 점수가 951,221 라고 하자

왼쪽 부분의 숫자의 합은 9 + 5 + 1 = 15 이다.

반대로 오른쪽 부분은 2 + 2 + 1 = 5 이다.

둘이 같으면 Lucky를, 아니면 Ready를 출력하도록 하자

```python
score = str(input())
```

```python
length = len(score)
left = score[:length//2]
right = score[length//2:]
l, r = 0, 0
for i,j in zip(left,right):
    l += int(i)
    r += int(j)
if l==r:
    print('Lucky')
else:
    print('Ready')
```

### 문자열 재정렬

> 조건

알파벳과 숫자로 이루어진 문자열이 입력된다.

알파벳은 정렬하고, 숫자는 모두 합한 뒤 정렬된 알파벳 뒤에 붙여서 출력한다.

> 예시

    입력 : K1KA3
    
    출력 : AKK4

```python
string = map(str, input())
alphabets, number = [], 0

for s in string:
    if ord('A')<=ord(s)<=ord('z'):
        alphabets.append(s)
    else:
        number += int(s)
alphabets.sort()

result = ''
for a in alphabets:
    result = result + a
result = result + str(number)

print(result)
```

### 도시 분할 계획

> Rule
 - N개의 집이 있고, M개의 도로들이 집들을 연결하고 있다.
 - R개의 마을로 분할한다. (단, N>R)
 - 도로 건설 비용을 최소로 하려고 할 때 드는 비용

> 입력
 - 첫번째 줄 : N, M, R
 - 두번째 줄~ : A, B, C
     - A와 B 집을 연결하는데 C의 비용이 필요  

```python
# graph = []
# N, M, R  = map(int, input().split())
# for _ in range(M):
  #   A, B, C = map(int, input().split())
  #   graph.append((C, min(A, B), max(A, B)))

N, M, R = 5, 7, 2
parent = list(range(N+1))
graph = [[2, 1, 2], [3, 1, 3], [4, 1, 4], [5, 1, 5], [6, 2, 3], [7, 2, 4], [2, 3, 5]]
```

```python
def find_parent(parent,i):
    if parent[i] != i:
        return find_parent(parent, parent[i])
    else:
        return parent[i]

def union_parent(parent, i, j):
    i, j = min(i, j), max(i, j)
    if parent[i] != parent[j]:
        parent[i] = j
```

```python
parent = list(range(N+1))
graph.sort()
result = 0


for i, g in enumerate(graph):
    if i>N-R-1:
        break
    cost, start, end = g
    if find_parent(parent, start) != find_parent(parent, end):
        union_parent(parent, start, end)
        result += cost
            
print(result)
```
