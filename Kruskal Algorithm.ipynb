# Kruskal Algorithm

    - 최소 비용을 가지는 Spanning Tree 만들기
    
    

### Example



```python
N = 7    # node의 개수
graph = [(10, 1, 2), (30, 1, 3), (15, 1, 7),(20, 2, 3),
         (25, 4, 7), (30, 4, 6), (10, 4, 5), (10, 5, 6)]    # (distance, start node, end node)
parent_tab = list(range(N+1))
```

```python
# 지난번 포스팅에서 사용한 함수들 그냥 복붙함

def find_parent(parent, i):
    if parent[i]!=i:
        parent[i] = find_parent(parent, parent[i])  
    return parent[i]

def union_parent(parent, i,j):
    i, j = find_parent(parent, i), find_parent(parent, j)
    i, j = max(i, j), min(i, j)
    parent[j] = i
```

```python
graph.sort()    # distance 순으로 정렬
result = []

for g in graph:
    d, i, j  = g
    if find_parent(parent_tab, i) == find_parent(parent_tab, j): # 사이클이 존재하면
        pass
    else:  
        union_parent(parent_tab, i, j)
        result.append((i, j , d))  
```

```python
total_dist = 0
for r in result:
    i, j, d = r
    total_dist += d
    print(f'{i} ~ {j} ---> {d}')
print(f'\n Total Distance : {total_dist}')
```
